# Algoritmo Genetico

Los Algoritmos Genéticos (AG) son un tipo de algoritmo de optimización y búsqueda inspirado en el proceso de selección natural. Aquí hay una descripción general rápida:

$\hspace{0.5cm}$

### 1. Concepto:

  Los algoritmos genéticos imitan el proceso de evolución natural. Se usan para desarrollar soluciones a los problemas de optimización y búsqueda.

$\hspace{0.5cm}$

### 2. Componentes:

  - **Población:** un conjunto de soluciones potenciales (a menudo representadas como cadenas binarias, pero no siempre).
  
  - **Función de aptitud:** Mide qué tan adecuada es una solución (qué tan buena es la solución para resolver el problema).
  
  - **Selección:** El proceso de seleccionar soluciones de la población para la reproducción. Las soluciones con mayor aptitud tienen mayores posibilidades de ser seleccionadas.
  
  - **Cruce (Recombinación)**: Dos soluciones seleccionadas (padres) producen uno o más descendientes, al combinar sus características.
  
  - **Mutación:** Introduce pequeños cambios aleatorios en la descendencia. Esto ayuda a mantener la diversidad genética y permite al AG explorar el espacio de búsqueda.

$\hspace{0.5cm}$

### 3. Proceso:

  - Inicializar una población de posibles soluciones.
  - Evaluar la aptitud de cada solución en la población.
  - Mientras no se cumpla una condición de terminación (como un número máximo de generaciones o se encuentre una solución con una idoneidad aceptable) se debe:
         a. Seleccionar soluciones para reproducción.
         b. Aplicar cruce y mutación para generar nuevas soluciones.
         C. Evaluar la idoneidad de las nuevas soluciones.
         d. Seleccione un conjunto de soluciones para la próxima generación.
  - Devolver la(s) mejor(es) solución(es).

$\hspace{0.5cm}$

### 4. Aplicaciones:
  - Optimización de funciones.
  - Entrenamiento de modelos por Machine Learning.
  - Desarrollo de estrategias de juego.
  - Problemas de programación/organizacion/optimizacion (NP-P)
  - Optimización del diseño.

$\hspace{0.5cm}$

### 5. Ventajas:

  Debido a que se trabaja en un gran espacio de posibles soluciones, no se requiere de ninguna información derivada.Se puede utilizar en problemas de optimización discretos, combinatorios y de variables mixtas.

$\hspace{0.5cm}$

### 6. Limitaciones:
  Dada la naturaleza del algoritmo, puede que no siempre se encuentre el óptimo global.Tambien puede crear un bucle, en el que la optimizacion de las generaciones cause un atasco en óptimos locales. Ademas, siempre se requiere ajustar parámetros como la tasa de mutación, la tasa de cruce, el tamaño de la población, etc.

# Definicion del Problema - Ejercicio 1.

Encontrar, haciendo uso del método de Algoritmos Geneticos (AG), la raíz real del polinomio $P(x) = 5x^5 - 3x^4 - x^3 - 5x^2 - x - 3$ en el intervalo $[0.5, 1]$, usando números reales, una población de 100 listas, un solo tipo de cruce y selección por clasificación. Dibujar un cuadro de Aptitud contra Generación.

## Desarrollo

Una vez analizado el problema, procedemos paso a paso:

$\hspace{0.5cm}$

**Inicialización:** inicializaremos una población de 100 listas binarias de números reales en el intervalo $[0,5, 1]$. Estos seran nuestros cromosomas.

$\hspace{0.5cm}$

**Función de aptitud:** La aptitud de una solución será inversamente proporcional al valor absoluto de $P(x)$. Esto significa que las soluciones más cercanas a la raíz tendrán una mayor aptitud ya que $P(x)$ estará más cerca de $0$.

$Aptitud(x) = \frac{1}{1+∣P(x)∣}$

$\hspace{0.5cm}$

**Selección:** Realizaremos la selección por clasificacion. Esto implica clasificar la población según su condición física y luego seleccionar a los padres entre los que tengan un mejor desempeño.

$\hspace{0.5cm}$

**Cruce:** usaremos un cruce de un solo punto para números reales. Las operaciones reales se basarán en bits. Para el cruce, los bits de un padre se pueden combinar con bits de otro. Para la mutación, se puede invertir una parte del cromosoma. Tomaremos dos padres $p_1$ y $p_2$. El hijo $c$ se puede generar como:

$c = α × p_1 + (1 - α) × p_2$

Donde $α$ es un número aleatorio entre 0 y 1.

$\hspace{0.5cm}$

**Terminación:** Nos detendremos después de un número determinado de generaciones o si la aptitud alcanza un umbral (lo que indica que estamos muy cerca de la raíz).

$\hspace{0.5cm}$

**Trazado:** usaremos la biblioteca `matplotlib` para trazar el cuadro de Aptitud (fitness) contra Generación.

```python
pip install matplotlib
```

```python
import random
import matplotlib.pyplot as plt

# Definimos nuestra funcion polinomial
def P(x):
    return 5*x**5 - 3*x**4 - x**3 - 5*x**2 - x - 3

# Funcion Decodificadora: ya que tenemos cadenas binarias, debemos convertirlas a un numero real en [0.5, 1]
def decode(chromosome):
    decimal_value = int(chromosome, 2)
    max_value = 2**len(chromosome) - 1
    return 0.5 + 0.5 * (decimal_value / max_value)

# Definimos nuestra funcion de aptitud
def fitness(chromosome):
    x = decode(chromosome)
    value = P(x)
    return -abs(value)

# Definimos como realizaremos los cruces
def crossover(parent1, parent2):
    index = random.randint(0, len(parent1) - 1)
    child1 = parent1[:index] + parent2[index:]
    child2 = parent2[:index] + parent1[index:]
    return child1, child2

# Funcion de Mutacion
def mutate(chromosome):
    index = random.randint(0, len(chromosome) - 1)
    mutated = list(chromosome)
    mutated[index] = '1' if mutated[index] == '0' else '0'
    return ''.join(mutated)

def main():
    # Parametros
    POP_SIZE = 200
    MAX_GEN = 1000
    MUTATION_RATE = 0.01
    CHROMOSOME_LENGTH = 100

    population = [''.join(random.choice(['0', '1']) for _ in range(CHROMOSOME_LENGTH)) for _ in range(POP_SIZE)]
    avg_fitnesses = [] #Aptitud Promedio

    #Analizamos las generaciones
    for generation in range(MAX_GEN):
        population.sort(key=fitness, reverse=True)
        avg_fitness = sum(fitness(chrom) for chrom in population) / POP_SIZE
        avg_fitnesses.append(avg_fitness)

        if generation % 100 == 0:
            best_x = decode(population[0])
            print(f"Generacion {generation}, Ap. Promedio: {avg_fitness}, Mejor x: {best_x}, Valor de P(x) = {P(best_x)}")

        new_population = population[:10]  # Elitismo: Nos quedamos con solo las mejores 10 soluciones


        #Mientras las generaciones tengan una buena actitud, creamos los padres.
        while len(new_population) < POP_SIZE:
            parent1 = random.choice(population[:50])
            parent2 = random.choice(population[:50])
            child1, child2 = crossover(parent1, parent2)

            if random.random() < MUTATION_RATE:
                child1 = mutate(child1)
            if random.random() < MUTATION_RATE:
                child2 = mutate(child2)

            new_population.extend([child1, child2])

        population = new_population

    #Se imprime la mejor solucion
    best_solution = decode(population[0])
    print(f"\nMejor Solucion hallada en: x = {best_solution}")
    print(f"Valor de P(x) en esta solucion: {P(best_solution)}")

    # Por ultimo, creamos el cuadro de Aptitud vs. Generacion
    plt.plot(avg_fitnesses)
    plt.xlabel('Generacion')
    plt.ylabel('Aptitud Promedio')
    plt.title('Aptitud vs. Generacion')
    plt.show()

main()
```

# Definicion del Problema - Ejercicio 2.

Supongamos que tenemos una lista de 15 alimentos con número de calorías, gramos de proteína, gramos de grasa y gramos de carbohidratos. Supongamos ademas que un individuo debe consumir determinada cantidad de calorías, proteínas, grasa y carbohidratos. Generar una dieta para ese individuo.



## Desarrollo

Tenemos una lista de 15 alimentos (escogidos de forma subjetiva y/o aleatoria). Cada alimento se caracteriza por su cantidad de calorías, gramos de proteína, gramos de grasa y gramos de carbohidratos.

Tenemos ademas un individuo con una ingesta objetivo de calorías, proteínas, grasas y carbohidratos.

Queremos utilizar un Algoritmo Genético para generar una dieta (combinación de los 15 alimentos) que satisfaga o se acerque más a satisfacer los requerimientos nutricionales del individuo.

$\hspace{0.5cm}$

**Representación:** Un cromosoma representará una dieta potencial. Será una lista de 15 números enteros, donde cada número entero representa el número de porciones de un alimento correspondiente.

$\hspace{0.5cm}$

**Función de Aptitud:** La diferencia entre el contenido nutricional total de la dieta y los requerimientos individuales. Cuanto menor sea la diferencia, mejor será la aptitud.

$\hspace{0.5cm}$

**Selección:** basada en rangos. Clasificaremos a la población según su condición física.

$\hspace{0.5cm}$

**Crossover:** Cruce de un punto. Dadas las dietas de dos padres, las combinaremos para generar dos dietas infantiles.

$\hspace{0.5cm}$

**Mutación:** cambia aleatoriamente la cantidad de porciones de un alimento aleatorio.

$\hspace{0.5cm}$

**Terminación:** Detener después de un número determinado de generaciones o si se encuentra una dieta con una aptitud lo suficientemente buena.

```python
import random

# Lista de los 15 alimentos con sus respectivos valores nutricionales
foods = [
    {"name": "Pollo", "calories": 165, "protein": 31, "fat": 3.6, "carbs": 0},
    {"name": "Carne", "calories": 332, "protein": 25, "fat": 27, "carbs": 0},
    {"name": "Huevos", "calories": 68, "protein": 5.5, "fat": 4.8, "carbs": 0.6},
    {"name": "Leche", "calories": 103, "protein": 8, "fat": 2.4, "carbs": 12},
    {"name": "Arroz", "calories": 205, "protein": 4.3, "fat": 0.4, "carbs": 45},
    {"name": "Pan", "calories": 79, "protein": 2.7, "fat": 0.97, "carbs": 15},
    {"name": "Papas", "calories": 130, "protein": 2, "fat": 0.2, "carbs": 30},
    {"name": "Manzanas", "calories": 95, "protein": 0.5, "fat": 0.3, "carbs": 25},
    {"name": "Bananas", "calories": 105, "protein": 1.3, "fat": 0.4, "carbs": 27},
    {"name": "Broccoli", "calories": 55, "protein": 3.7, "fat": 0.6, "carbs": 11},
    {"name": "Espinaca", "calories": 23, "protein": 2.9, "fat": 0.4, "carbs": 3.6},
    {"name": "Queso", "calories": 113, "protein": 7, "fat": 9, "carbs": 0.4},
    {"name": "Almendras", "calories": 164, "protein": 6, "fat": 14, "carbs": 6},
    {"name": "Yoghurt", "calories": 154, "protein": 13, "fat": 8.5, "carbs": 11},
    {"name": "Maiz", "calories": 96, "protein": 3.2, "fat": 1.5, "carbs": 21},
]


# Requisitos nutricionales del individuo (subjetivos)
requirements = {
    "calories": 2500,
    "protein": 150,
    "fat": 50,
    "carbs": 300
}

# Calcular el contenido nutricional total de una dieta.
def total_nutrition(diet):
    total = {"calories": 0, "protein": 0, "fat": 0, "carbs": 0}
    for i, servings in enumerate(diet):
        for nutrient in total:
            total[nutrient] += foods[i][nutrient] * servings
    return total

# Funcion de Aptitud
def fitness(diet):
    nutrition = total_nutrition(diet)
    difference = sum(abs(nutrition[n] - requirements[n]) for n in requirements)
    return -difference  # Queremos maximizar la aptitud, pero minimizar la diferencia.

# Funcion de Cruce
def crossover(parent1, parent2):
    index = random.randint(0, len(parent1) - 1)
    child1 = parent1[:index] + parent2[index:]
    child2 = parent2[:index] + parent1[index:]
    return child1, child2

# Funcion de Mutacion
def mutate(diet):
    index = random.randint(0, len(diet) - 1)
    diet[index] += random.randint(-1, 1)
    diet[index] = max(0, diet[index])  # Nos aseguramos que no se den valores negativos (no tienen sentido con alimentos)
    return diet

def main():
    POP_SIZE = 100
    MAX_GEN = 1000
    population = [[random.randint(0, 5) for _ in range(15)] for _ in range(POP_SIZE)]

    for generation in range(MAX_GEN):
        population.sort(key=fitness, reverse=True)

        if generation % 100 == 0:
            print(f"Generacion {generation}, Mejor Aptitud: {fitness(population[0])}")

        new_population = population[:10]  # Elitismo: Nos interesan las mejores 10 dietas

        while len(new_population) < POP_SIZE:
            parent1 = random.choice(population[:50])
            parent2 = random.choice(population[:50])
            child1, child2 = crossover(parent1, parent2)
            new_population.append(mutate(child1))
            new_population.append(mutate(child2))

        population = new_population

    best_diet = population[0]
    print("\nMejor dieta hallada:")
    for i, servings in enumerate(best_diet):
        print(f"{foods[i]['name']}: {servings} Porciones")

main()
```
